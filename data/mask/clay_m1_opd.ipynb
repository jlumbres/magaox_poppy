```python
import numpy as np
from astropy import units as u
from astropy.io import fits
from matplotlib import pyplot as plt
```

```python
clay = fits.open('ClayM1_0mask_meters_new.fits')[0]
surf = clay.data
hdr = clay.header
hdr
```

```python
opd = surf * 2
fits.writeto('ClayM1_0mask_meters_opd.fits', opd, hdr, overwrite=True)
```

```python
plt.figure(dpi=100)
plt.imshow(opd, origin='lower')
plt.colorbar().set_label('meters')
plt.title('Magellan Clay M1 (OPD)')
```
